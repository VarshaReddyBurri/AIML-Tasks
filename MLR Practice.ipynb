```python
import pandas as pd
import numpy as np
import seaborn as sns
import statsmodels.formula.api as smf
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the data
# from google.colab import files
# uploaded = files.upload()

cars=pd.read_csv("Toyoto_COrrola.csv")
cars.head()
```

```python
cars.info()
```

```python
# Checking duplicates
cars[cars.duplicated()]
```

```python
cars.isnull().sum()
```

#### Observations
- There are no null values
- There are no duplicated values
- The Continuos are PRICE, AGE, KM, HP, WEIGHT.
- The categorical datatype are DOORS, CyCLINDERS, GEARS

```python
fig, axes = plt.subplots(2,1, figsize=(8,6), gridspec_kw={'height_ratios':[1,3]})
sns.boxplot(data=cars["Price"], ax=axes[0], color='skyblue', width=0.5, orient='h')
axes[0].set_title("Boxplot")
axes[0].set_xlabel("Price Levels")

sns.histplot(cars["Price"], kde=True, ax=axes[1], color="purple", bins=30)
axes[1].set_title("Histogram with KDE")
axes[1].set_xlabel("Price Levels")
axes[1].set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```
